```python
def linear_func(B, x):
    return B[0] * x + B[1]

def get_odr(x, y):
    '''
    orthogonal distance regression to evaluate pred and obs
    '''
    
    model = odr.Model(linear_func)
    data = odr.RealData(x, y)
    od = odr.ODR(data, model, beta0=[1., 0.])
    out = od.run()
    
    slope, intercept = out.beta
    # Correlation
    r = np.corrcoef(x, y)[0, 1]
    return slope, intercept, r
```

```python
def reg(*args):
    """
    Multi-regression returning R^2, coefficients, intercept.
    Handles 2–4 predictors; filters NaNs automatically.
    """
    # Split predictors and target
    *predictors, y = [np.asarray(a).flatten() for a in args]

    # Column-stack predictors dynamically
    X = np.column_stack(predictors)

    # Build a single valid mask for all variables
    mask = ~np.isnan(X).any(axis=1) & ~np.isnan(y)

    X = X[mask]
    y = y[mask].reshape(-1, 1)

    regressor.fit(X, y)
    return regressor.score(X, y), regressor.coef_[0], regressor.intercept_
```

```python
def binbyrank(var, nbin):
    '''
    return the bin indices; indices of the original variable
    var shape: cycnum
    '''
    bins = []
    binsize=len(var)//nbin
    sortedvar = var.copy()
    sortedvar.sort()
    # 50 bins; -0.02-0.02, 0.0008
    for i in range(0, nbin):
        low = sortedvar[i*binsize]
        up = sortedvar[(i+1)*binsize-1]
        idc = np.where((var>=low)&(var<=up))[0]
        bins.append(idc)
    return bins
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

```python
# convert back based on SAA
def convert_alb(albcorr, SAA, a=0.6, b=0.4):
    '''
    convert SAA corrected alb to actual alb
    alb and SAA: array of same length
    a,b: param to fit SAA/alb function
    '''
    sinSAA = np.sin(np.deg2rad(SAA))
    k = (albcorr-a)/1   # dx: sin(90)**b-0 = 1
    actualAlb = k*sinSAA**b + a
    return actualAlb

def corr_alb_SAA(alb, SAA, a=0.6, b=0.4):
    '''
    correct albedo by solar altitude angle
    following Cronin 2014
    '''
    return (a-alb)/(-np.sin(np.deg2rad(SAA))**b)*1+a
```

# NN

```python
class ShallowMLP(nn.Module):
    """
    Shallow neural network for nonlinear regression:
    X (meteorology) -> y (cloud property)

    Architecture:
        input -> linear -> tanh -> linear -> output
    """
    def __init__(self, input_dim, hidden_dim):
        super().__init__()

        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.activation = nn.Tanh()
        self.fc2 = nn.Linear(hidden_dim, 1)


    def forward(self, x):
        """
        x: tensor of shape (batch_size, input_dim)
        """
        h = self.activation(self.fc1(x))
        y_hat = self.fc2(h)
        return y_hat.squeeze(-1)
```

```python
def train_model(
    model,
    train_loader,
    val_loader,
    optimizer,
    criterion,
    max_epochs=500,
    patience=20,
):
    best_val_loss = float("inf")
    best_state = model.state_dict()
    patience_counter = 0

    for epoch in range(max_epochs):
        # ---- training ----
        model.train()
        train_loss = 0.0
        
        for x_batch, y_batch in train_loader:
            optimizer.zero_grad()
            y_pred = model(x_batch)
            loss = criterion(y_pred, y_batch)
            loss.backward()
            # torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)  # debugging
            optimizer.step()
            train_loss += loss.item()

        train_loss /= len(train_loader)

        # ---- validation ----
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for x_batch, y_batch in val_loader:
                y_pred = model(x_batch)
                loss = criterion(y_pred, y_batch)
                val_loss += loss.item()

        val_loss /= len(val_loader)

        if not np.isfinite(val_loss):
            raise ValueError("Validation loss is NaN or inf")

        # ---- early stopping ----
        if val_loss <= best_val_loss:
            best_val_loss = val_loss
            best_state = model.state_dict()
            patience_counter = 0
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f"Early stopping at epoch {epoch}")
                break

        if epoch % 20 == 0:
            print(
                f"Epoch {epoch:4d} | "
                f"Train loss: {train_loss:.4f} | "
                f"Val loss: {val_loss:.4f}"
            )
        
        # debugging
        # print(f"Epoch {epoch}, val_loss = {val_loss}")

    
    model.load_state_dict(best_state)
    return model
```

```python
class nnDataset(Dataset):
    def __init__(self, X, y):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
def partial_dependence(model, X_ref, feature_index, grid_points=50):
    """
    Computes partial dependence for one feature.

    X_ref: reference dataset (N, D), standardized
    feature_index: index of variable to vary
    """
    X_mean = X_ref.mean(axis=0)

    x_vals = np.linspace(
        X_ref[:, feature_index].min(),
        X_ref[:, feature_index].max(),
        grid_points
    )

    y_vals = []

    model.eval()
    with torch.no_grad():
        for v in x_vals:
            X_tmp = X_mean.copy()     # holding other X constant as mean
            X_tmp[feature_index] = v    # varying the value of target X
            X_tmp = torch.tensor(X_tmp[None, :], dtype=torch.float32)
            y_hat = model(X_tmp).item()   # modeled y with mean state + perturbed X[j]
            y_vals.append(y_hat)

    return x_vals, np.array(y_vals)
```

```python
def compute_sensitivities(model, X):
    """
    Computes dy/dX for each sample.
    X: (N, D) NumPy array (standardized)
    Returns: (N, D) sensitivities
    """
    X_t = torch.tensor(X, dtype=torch.float32, requires_grad=True)
    model.eval()

    y = model(X_t)           # (N,)
    y.sum().backward()       # accumulate gradients

    grads = X_t.grad.detach().numpy()
    return grads
```
